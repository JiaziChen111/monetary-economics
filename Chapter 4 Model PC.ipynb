In the models below, the variable *R* is defined as a parameter and not a variable.  This is due to the structure of the equations, if R was a variable, then this would not longer be a set of linear equations, thus I treat R as a parameter.

### Model PC

```python
from pylinsolve.model import Model
from pylinsolve.utils import is_close,round_solution

def create_pc_model():
    model = Model()

    model.set_var_default(0)
    model.var('Bcb', desc='Government bills held by the Central Bank')
    model.var('Bh', desc='Government bills held by households')
    model.var('Bs', desc='Government bills supplied by the government')
    model.var('C', desc='Consumption goods')
    model.var('Hh', desc='Cash held by households')
    model.var('Hs', desc='Cash supplied by the central bank')
    #model.var('R', desc='Interest rate on government bills')
    model.var('T', desc='Taxes')
    model.var('V', desc='Household wealth')
    model.var('Y', desc='Income = GDP')
    model.var('YD', desc='Disposable income of households')

    model.param('alpha1', desc='Propensity to consume out of income', default=0.6)
    model.param('alpha2', desc='Propensity to consume o of wealth', default=0.4)
    model.param('lambda0', desc='Parameter in asset demand function', default=0.635)
    model.param('lambda1', desc='Parameter in asset demand function', default=5.0)
    model.param('lambda2', desc='Parameter in asset demand function', default=0.01)
    model.param('theta', desc='Tax rate', default=0.2)

    model.param('G', desc='Government goods', default=20.)
    model.param('R', desc='Interest rate on government bills', default=0.025)
    # model.param('Rbar', desc='Interest rate as policy instrument', default=0.025)

    model.add('Y = C + G')  # 4.1
    model.add('YD = Y - T + R(-1)*Bh(-1)')  # 4.2
    model.add('T = theta*(Y + R(-1)*Bh(-1))')  #4.3, theta < 1
    model.add('V = V(-1) + (YD - C)')  # 4.4
    model.add('C = alpha1*YD + alpha2*V(-1)')  # 4.5, 0<alpha2<alpha1<1
    model.add('Hh = V - Bh')  # 4.6
    model.add('Bh = V*lambda0 + V*lambda1*R - lambda2*YD') # 4.7
    model.add('Bs - Bs(-1) = (G + R(-1)*Bs(-1)) - (T + R(-1)*Bcb(-1))')  # 4.8
    model.add('Hs - Hs(-1) = Bcb - Bcb(-1)')  # 4.9
    model.add('Bcb = Bs - Bh')  # 4.10
    # model.add('R = Rbar')  # 4.11
        
    return model
```

```python
steady = create_pc_model()
for _ in xrange(100):
    steady.solve(iterations=100, threshold=1e-5)
    if is_close(steady.solutions[-2], steady.solutions[-1], atol=1e-4):
        break
```

### Model PCEX

```python
from pylinsolve.model import Model
from pylinsolve.utils import is_close,round_solution

def create_pcex_model():
    model = Model()

    model.set_var_default(0)
    model.var('Bcb', desc='Government bills held by the Central Bank', default=116.36)
    model.var('Bd', desc='Demand for government bills')
    model.var('Bh', desc='Government bills held by households', default=363.59)
    model.var('Bs', desc='Government bills supplied by the government', default=479.95)
    model.var('C', desc='Consumption goods')
    model.var('Hd', desc='Demand for cash')
    model.var('Hh', desc='Cash held by households', default=116.35)
    model.var('Hs', desc='Cash supplied by the central bank', default=116.35)
    model.var('T', desc='Taxes')
    model.var('V', desc='Household wealth', default=479.94)
    model.var('Ve', desc='Expected household wealth')
    model.var('Y', desc='Income = GDP')
    model.var('YD', desc='Disposable income of households')
    model.var('YDe', desc='Expected disposable income of households')

    model.set_param_default(0)
    model.param('alpha1', desc='Propensity to consume out of income', default=0.6)
    model.param('alpha2', desc='Propensity to consume o of wealth', default=0.4)
    model.param('lambda0', desc='Parameter in asset demand function', default=0.635)
    model.param('lambda1', desc='Parameter in asset demand function', default=5.0)
    model.param('lambda2', desc='Parameter in asset demand function', default=0.01)
    model.param('theta', desc='Tax rate', default=0.2)

    model.param('G', desc='Government goods', default=20.)
    model.param('R', desc='Interest rate on government bills', default=0.025)
    model.param('Ra', desc='Random shock to expectations', default=0.0)
    # model.param('Rbar', desc='Interest rate as policy instrument', default=0.025)

    model.add('Y = C + G')  # 4.1
    model.add('YD = Y - T + R(-1)*Bh(-1)')  # 4.2
    model.add('T = theta*(Y + R(-1)*Bh(-1))')  #4.3, theta < 1
    model.add('V = V(-1) + (YD - C)')  # 4.4
    model.add('C = alpha1*YDe + alpha2*V(-1)')  # 4.5E
    model.add('Bd = Ve*lambda0 + Ve*lambda1*R - lambda2*YDe')  # 4.7E
    model.add('Hd = Ve - Bd')  # 4.13
    model.add('Ve = V(-1) + (YDe - C)')  # 4.14
    model.add('Hh = V - Bh')  # 4.6
    model.add('Bh = Bd')  # 4.15
    model.add('Bs - Bs(-1) = (G + R(-1)*Bs(-1)) - (T + R(-1)*Bcb(-1))')  # 4.8
    model.add('Hs - Hs(-1) = Bcb - Bcb(-1)')  # 4.9
    model.add('Bcb = Bs - Bh')  # 4.10
    # model.add('R = Rbar')  # 4.11
    model.add('YDe = YD * (1 + Ra)')  # 4.16
        
    return model
```

### Steady state and shocks

```python
pcex_steady = create_pcex_model()
for _ in xrange(100):
    pcex_steady.solve(iterations=100, threshold=1e-5)

    if is_close(pcex_steady.solutions[-2], pcex_steady.solutions[-1], atol=1e-4):
        break
```

```python
import random

random.seed(6)

shocks = create_pcex_model()
shocks.set_variables(pcex_steady.solutions[0])
for _ in xrange(50):
    shocks.parameters['Ra'].value = random.gauss(0,1) / 10.
    shocks.solve(iterations=100, threshold=1e-5)
```

### Steady state with increase in interest rate

```python
rate_shock = create_pc_model()
rate_shock.variables['Bcb'].value = 21.576
rate_shock.variables['Bh'].value = 64.865
rate_shock.variables['Bs'].value = 86.441
rate_shock.variables['Hh'].value = 21.62
rate_shock.variables['Hs'].value = 21.62
rate_shock.variables['V'].value = 86.485

# solve until stable
for i in xrange(50):
    rate_shock.solve(iterations=100, threshold=1e-5)
    if is_close(rate_shock.solutions[-2], rate_shock.solutions[-1], atol=1e-4):
        break

for i in xrange(30):
    rate_shock.solve(iterations=100, threshold=1e-5)
    if i == 5:
        rate_shock.parameters['R'].value = 0.035
```

### Model PCEX2

In the code below, *alpha1* is treated as a parameter and not a variable.  Due to the way the solver is implemented, a linear system is required.  Due to equation 4.7E, the two variables cannot be mixed, thus I treat it as a parameter rather than a true variable.

```python
from pylinsolve.model import Model

def create_pcex2_model():
    model = Model()

    model.set_var_default(0)
    model.var('Bcb', desc='Government bills held by the Central Bank', default=21.576)
    model.var('Bd', desc='Demand for government bills')
    model.var('Bh', desc='Government bills held by households', default=64.865)
    model.var('Bs', desc='Government bills supplied by the government', default=86.441)
    model.var('C', desc='Consumption goods')
    model.var('Hd', desc='Demand for cash')
    model.var('Hh', desc='Cash held by households', default=21.62)
    model.var('Hs', desc='Cash supplied by the central bank', default=21.62)
    model.var('T', desc='Taxes')
    model.var('V', desc='Household wealth', default=86.485)
    model.var('Ve', desc='Expected household wealth')
    model.var('Y', desc='Income = GDP')
    model.var('YD', desc='Disposable income of households', default=90)
    model.var('YDe', desc='Expected disposable income of households')

    model.set_param_default(0)
    model.param('alpha1', desc='Propensity to consume out of income', default=0.6)
    model.param('alpha2', desc='Propensity to consume out of wealth', default=0.4)
    model.param('alpha10', desc='Propensity to consume out of income - exogenous', default=0.7)
    model.param('iota', desc='Impact of interest rate on the propensity to consume out of income', default= 4)
    model.param('lambda0', desc='Parameter in asset demand function', default=0.635)
    model.param('lambda1', desc='Parameter in asset demand function', default=5.0)
    model.param('lambda2', desc='Parameter in asset demand function', default=0.01)
    model.param('theta', desc='Tax rate', default=0.2)

    model.param('G', desc='Government goods', default=20.)
    model.param('R', desc='Interest rate on government bills', default=0.025)
    # model.param('Ra', desc='Random shock to expectations', default=0.0)
    # model.param('Rbar', desc='Interest rate as policy instrument', default=0.025)

    model.add('Y = C + G')  # 4.1
    model.add('YD = Y - T + R(-1)*Bh(-1)')  # 4.2
    model.add('T = theta*(Y + R(-1)*Bh(-1))')  #4.3, theta < 1
    model.add('V = V(-1) + (YD - C)')  # 4.4
    model.add('C = alpha1*YDe + alpha2*V(-1)')  # 4.5E
    model.add('Bd = Ve*lambda0 + Ve*lambda1*R - lambda2*YDe')  # 4.7E
    model.add('Hd = Ve - Bd')  # 4.13
    model.add('Ve = V(-1) + (YDe - C)')  # 4.14
    model.add('Hh = V - Bh')  # 4.6
    model.add('Bh = Bd')  # 4.15
    model.add('Bs - Bs(-1) = (G + R(-1)*Bs(-1)) - (T + R(-1)*Bcb(-1))')  # 4.8
    model.add('Hs - Hs(-1) = Bcb - Bcb(-1)')  # 4.9
    model.add('Bcb = Bs - Bh')  # 4.10
    # model.add('R = Rbar')  # 4.11
    model.add('YDe = YD(-1)')  # 4.16A
    # model.add('alpha1 = alpha10 - iota*R(-1)')
        
    return model
```

```python
from pylinsolve.utils import is_close,round_solution

pcex2 = create_pcex2_model()
for i in xrange(10):
    pcex2.solve(iterations=100, threshold=1e-5)

for i in xrange(40):
    if i == 1:
        pcex2.parameters['R'].value += 0.01
    pcex2.parameters['alpha1'].value = pcex2.evaluate('alpha10 - iota*R(-1)')
    pcex2.solve(iterations=100, threshold=1e-5)
```
